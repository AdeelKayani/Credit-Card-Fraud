```
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization styles
sns.set(style="whitegrid")

# Upload the dataset manually in Colab
from google.colab import files
data = pd.read_csv("/content/creditcard.csv")

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# 1. How many rows and columns are in the dataset?
print("\nShape of the dataset:")
print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")

# 2. What are the column names and their data types?
print("\nColumn names and data types:")
print(data.dtypes)

# 3. Are there any missing or null values in the dataset?
print("\nMissing or null values in the dataset:")
print(data.isnull().sum())

# 4. How many transactions are fraudulent, and how many are legitimate?
print("\nCount of fraudulent and legitimate transactions:")
fraud_counts = data['Class'].value_counts()
print(fraud_counts)

# 5. What percentage of transactions are fraudulent?
fraud_percentage = (fraud_counts[1] / fraud_counts.sum()) * 100
print(f"\nPercentage of fraudulent transactions: {fraud_percentage:.2f}%")

# 6. Descriptive statistics for numerical columns like 'Amount'
print("\nDescriptive statistics for the 'Amount' column:")
print(data['Amount'].describe())

# 7. What is the maximum transaction amount in the dataset, and is it fraudulent?
max_transaction = data.loc[data['Amount'].idxmax()]
print("\nMaximum transaction amount and details:")
print(max_transaction)

# 8. Bar chart showing the count of fraudulent vs. legitimate transactions
plt.figure(figsize=(8, 6))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, palette="viridis")
plt.title("Count of Fraudulent vs. Legitimate Transactions")
plt.xticks([0, 1], ['Legitimate', 'Fraudulent'])
plt.ylabel("Count")
plt.xlabel("Transaction Type")
plt.show()

# 9. Histogram of transaction amounts
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount'], bins=50, kde=True, color='blue')
plt.title("Histogram of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()

# 10. Heatmap to visualize the correlation between numerical features
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Between Features")
plt.show()
```
